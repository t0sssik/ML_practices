1) Разработать модель, которая предскажет вероятность снижения покупательской активности.

### Фильтрация покупателей

Прежде чем проводить исследовательский анализ данных, отфильтруем необходимых покупателей, у которых наблюдается активность в течение не менее трех месяцев. Для этого посмотрим на данные из таблицы `market_money`

```python
grouped_cust = market_money.pivot_table(index='id', columns='Период', values='Выручка').reset_index()
grouped_cust.head()
```

```python
active_customers = grouped_cust[
    (grouped_cust['предыдущий_месяц'] > 0) & 
    (grouped_cust['препредыдущий_месяц'] > 0) & 
    (grouped_cust['текущий_месяц'] > 0)
]['id'].to_list()
len(active_customers)
```

Получается, что из всех покупателей только `3` не подходят для анализа. Применим фильтр ко всем таблицам

```python
market_file = market_file.query('id in @active_customers')
market_file.info()
```

```python
market_money = market_money.query('id in @active_customers')
market_money.info()
```

```python
market_time = market_time.query('id in @active_customers')
market_time.info()
```

```python
money = money.query('id in @active_customers')
money.info()
```

## Обучение модели

### Обучение без добавления новых признаков (не считая кодирования)

Для более точного подбора способов подготовки данных, моделей и гиперпараметров воспользуемся пайплайнами. Они позволят легко подобрать лучшие "переменные" для поставленной задачи

Также, необходимо выбрать метрику, которая будет оценивать качество моделей. Так как классы несбалансированы, то метрика `accuracy` точно не подойдет. В качестве основной метрики можно взять `ROC-AUC` либо `F1 score`. ROC-AUC более устойчива при работе с несбалансированными данными. Также, нам важно оценить способность модели различать между классами независимо от порога классификации. поэтому остановимся именно на этой метрике

Сначала разделим данные для обучения на тренировочную и тестовую выборки. Так как есть дисбаланс, то будем использовать параметр `stratify`

```python
RANDOM_STATE = 42
TEST_SIZE = 0.25

df.set_index('id', inplace=True)
X = df.drop('Покупательская активность', axis=1)
y = df['Покупательская активность']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y
)
```

Выполним кодирование целевого признака с помощью `Label Encoder`

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)

y_train = label_encoder.transform(y_train)
y_test = label_encoder.transform(y_test) 
```

```python
ohe_columns = ['Популярная_категория', 'Тип сервиса', 'Разрешить сообщать']

# ord_columns = ['Тип сервиса', 'Разрешить сообщать']

num_columns = [
    'Маркет_актив_6_мес',
    'Маркет_актив_тек_мес',
    'Длительность',
    'Акционные_покупки',
    'Средний_просмотр_категорий_за_визит',
    'Неоплаченные_продукты_штук_квартал',
    'Ошибка_сервиса',
    'Страниц_за_визит',
    'Выручка_предыдущий_месяц',
    'Выручка_препредыдущий_месяц',
    'Выручка_текущий_месяц',
    'Посещения_предыдущий_месяц',
    'Посещения_текущий_месяц'
]
```

```python
ohe_p = Pipeline(
    [
        ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ohe', OneHotEncoder(drop='first'))
    ]
)

data_preprocessor = ColumnTransformer(
    [('ohe', ohe_p, ohe_columns),
     ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
)

pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)

param_grid = [
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__min_samples_split': range(2, 5),
        'models__min_samples_leaf': range(1, 5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    },
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(1, 11),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    },
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', penalty='l1')],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    },
    {
        'models': [SVC(random_state=RANDOM_STATE, kernel='rbf', probability=True)],
        'models__degree': range(3, 11),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    }
]
```

В качестве метода для подбора лучших парамтеров будем использовать `GridSearchCV`, чтобы гарантировать лучший вариант

```python
grid_search = GridSearchCV(
    pipe_final,
    param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)
grid_search.fit(X_train, y_train)
```

```python
print('Лучшая модель и её параметры:\n\n', grid_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', grid_search.best_score_)
```

Лучшей моделью оказалась SVC, а метрика ROC AUC на тренировочных даных - 0.911. Теперь посмотрим на метрику на тестовых данных

```python
y_pred = grid_search.predict_proba(X_test)[:, 1]
print(f'Метрика ROC-AUC на тестовой выборке: {roc_auc_score(y_test, y_pred)}')
```

На тестовых данных результат получился даже чуть лучше. Можно сказать точно, что это лучший вариант из тех возможных, которые перебирались в пайплайне

```python
RANDOM_STATE = 42
TEST_SIZE = 0.25

X = df.drop('Покупательская активность', axis=1)
y = df['Покупательская активность']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y
)
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)

y_train = label_encoder.transform(y_train)
y_test = label_encoder.transform(y_test) 
```

```python
ohe_columns = ['Популярная_категория', 'Тип сервиса', 'Разрешить сообщать']

num_columns = [
    'Маркет_актив_6_мес',
    'Маркет_актив_тек_мес',
    'Длительность',
    'Акционные_покупки',
    'Средний_просмотр_категорий_за_визит',
    'Неоплаченные_продукты_штук_квартал',
    'Ошибка_сервиса',
    'Страниц_за_визит',
    'Выручка_предыдущий_месяц',
    'Выручка_препредыдущий_месяц',
    'Выручка_текущий_месяц',
    'Посещения_предыдущий_месяц',
    'Посещения_текущий_месяц',
    'Конверсия_текущий_месяц',
    'Конверсия_предыдущий_месяц',
    'Сравнительная_выручка',
    'Коэффициент_активности',
    'Индекс_ошибок'
]
```

```python
ohe_p = Pipeline(
    [
        ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ohe', OneHotEncoder(drop='first'))
    ]
)

data_preprocessor = ColumnTransformer(
    [('ohe', ohe_p, ohe_columns),
     ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
)

pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)

param_grid = [
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__min_samples_split': range(2, 5),
        'models__min_samples_leaf': range(1, 5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    },
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(1, 11),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    },
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', penalty='l1')],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    },
    {
        'models': [SVC(random_state=RANDOM_STATE, kernel='rbf', probability=True)],
        'models__degree': range(3, 11),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']
    }
]
```

```python
grid_search = GridSearchCV(
    pipe_final,
    param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)
grid_search.fit(X_train, y_train)
```

```python
print('Лучшая модель и её параметры:\n\n', grid_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', grid_search.best_score_)
```

```python
y_pred = grid_search.predict_proba(X_test)[:, 1]
print(f'Метрика ROC-AUC на тестовой выборке: {roc_auc_score(y_test, y_pred)}')
```
